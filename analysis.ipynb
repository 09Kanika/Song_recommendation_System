```python
def rating(a):
    return float(a[:-3])
```

```python
rating('9.6/10')
```

```python
data['User-Rating'].unique()
```

```python
data['User-Rating'].replace({'10-Oct': '7.0/10', "09-Oct":"7.0/10"}, inplace=True)
```

```python
data["User-Rating"] = data["User-Rating"].apply(rating)
```

```python
data.sample()
```

```python
data["Genre"].unique()
```

```python
data["interest"]=data['Genre'] +"^"+ data['Singer/Artists']+"^"+data['Album/Movie']
data.sample()
```

```python
def splitter(s):
    s = s.lower()
    return s.split('^')
```

```python
data['interest']=data['interest'].apply(splitter)
```

```python
data.sample()
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(lst):
    word = []

    for i in lst:
        word.append(ps.stem(i))
    
    return " ".join(word)
```

```python
data["interest"] = data["interest"].apply(stem)
data.sample()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors = cv.fit_transform(data["interest"]).toarray()
```

```python
vectors.shape
```

```python
vectors
```

```python
cv.get_stop_words()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
similarity.shape
```

```python
sorted(list(enumerate(similarity[0])), reverse=True, key=lambda x: x[1])
```

```python
data['Song-Name'].iloc[460]
```

```python
def recommended(song):
    idx = data[data["Song-Name"] == song].index[0]
    corr = similarity[idx]
    rec = sorted(list(enumerate(corr)), reverse=True, key=lambda x: x[1])[1:16]

    for i in rec:
        print(data['Song-Name'].iloc[i[0]])
```

```python
recommended('Main Kya Hoon')
```
